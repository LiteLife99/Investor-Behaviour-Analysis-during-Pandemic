## Testing for one dataframe successful
- now loop and append

```python
df2 = pd.DataFrame()
```

```python
symbols
```

```python
# adding Return and mean_return for the companies
%%capture
for i in range(0,len(symbols)):
    sym = symbols[i]
    df_test = df.query('Symbol in @sym').copy()
    print(df_test)
    df_test['Return'] = np.zeros((len(df_test['VWAP']),1)) #remember that dtype is int, not float
    df_test.Return = df_test.Return.astype(float)
    print(df_test['VWAP'].iloc[1])
    df_test.reset_index(drop=True, inplace=True)
    for j in range(0,len(df_test['VWAP'])-1):
        val = float((df_test['VWAP'].iloc[j+1]-df_test['VWAP'].iloc[j])/df_test['VWAP'].iloc[j]*100)
        df_test['Return'].iloc[j+1] = val
    df_test['mean_return_company'] = np.zeros((len(df_test['VWAP']),1)) #remember that dtype is int, not float
    df_test.mean_return_company = df_test.mean_return_company.astype(float)
    df_test['mean_return_company'] = df_test.Return.mean()
    print(df_test)
    df2 = df2.append(df_test,ignore_index=True)
```
